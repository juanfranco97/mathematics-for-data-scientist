# Vector

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def disp_vectors(vecs,cols,alpha=1):
    plt.figure()
    plt.axvline(x=0, color = "grey", zorder = 0)
    plt.axhline(y=0, color = "grey", zorder = 0)
    
    for i in range(len(vecs)):
        x= np.concatenate([[0,0], vecs[i]])
        plt.quiver([x[0]],
                   [x[1]],
                   [x[2]],
                   [x[3]],
                  angles='xy', scale_units='xy', scale=1,
                  color=cols[i],
                  alpha=alpha)
```

```python
v1 = np.array([2,5])
v2 = np.array([3,2])
disp_vectors([v1, v2], ['orange', 'blue'])
plt.xlim(-1, 8)
plt.ylim(-1, 8)
```

## vector linear combination

```python
v1=np.array([2,5])
v2 = np.array([3,2])
v1v2= v1+v2
```

```python
disp_vectors([v1, v2, v1v2], ['orange', 'blue', 'red'])
plt.xlim(-1, 8)
plt.ylim(-1, 12)
```

## linearly dependent vectors

 Two vectors are linearly dependent if area between it are equal to 0, in other words are superimposed.

```python
v1=np.array([2,5])
v2 = np.array([2,5])
for a in range(-10,10):
    for b in range(-10,10):
        plt.scatter(v1[0]*a + v2[0]*b, v1[1]*a + v2[1]*b,
                   marker= '.',
                   color = 'orange')
plt.xlim(-100,100)
plt.ylim(-100,100)
plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')

plt.show()
```

## norm 

V = vector
\begin{eqnarray}
norm(V) \ge 0 \\ 
norm(V) = 0 \leftrightarrow V= 0 \\
norm(V3) \le  norm(V1) + norm(V2) \\
norm(aV) = abs(a) \cdot norm(V)
\end{eqnarray}

```python
v1 = np.array([2,7])
v2 = np.array([3,5])
v1v2 = v1+v2
```

```python
norma_v1 = np.linalg.norm(v1)
norma_v2 = np.linalg.norm(v2)
norma_v1v2 = np.linalg.norm(v1v2)
print(norma_v1v2 <= (norma_v1 + norma_v2))
```

## norm types

### norm L0 :
Return how many elements are diferent to 0 

```python
#norm 0
vector = np.array([1,2,0,5,6,0])
print(np.linalg.norm(vector, ord=0))
```

### norm L1:
Resturn sum of elements absolute values 

```python
#norm 1 
vector = np.array([1,-1,1,-1,1])
print(np.linalg.norm(vector, ord=1))
```

### norm L2:
Return magnitude of the vector from its origin

```python
#norm 2 
vector = np.array([1,1])
print(np.linalg.norm(vector))
print(np.linalg.norm(vector, ord=2))
```

### norm $\infty$
returns the largest absolute value of the vector

```python
#norm infinito
vector = np.array([1,2,3,-100])
print(np.linalg.norm(vector, ord=np.inf))
```

## orthogonal vectors

the orthogonal vectors are those that form an angle of 90 °

```python
x = np.array([0,0,2,2])
y = np.array([0,0,2,-2])
plt.quiver([x[0],y[0]],
           [x[1],y[1]],
           [x[2],y[2]],
           [x[3],y[3]],
          angles = 'xy', scale_units = 'xy', scale=1)
plt.xlim(-2,4)
plt.ylim(-3,3)
plt.show()
```

```python
v1= np.array([2,2])
v2= np.array([2,-2])
print(v1.dot(v2.T))
```

## orthonormal vector
A vector is orthonormal if norm L2 == 1

```python
v1= np.array([2,2])
v2= np.array([2,-2])
n1 = np.linalg.norm(v1)
n2 = np.linalg.norm(v2)
print(n1) 
print(n2)
```

Not are orthonormal vectors

### convert a not orthonormal vector  to orthonormal

```python
newv1 = v1/n1
newv2 = v2/n2
print(np.linalg.norm(newv1)) 
print(np.linalg.norm(newv2))
```
